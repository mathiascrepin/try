```
import matplotlib.pyplot as plt
import numpy as np
```

```
# Données à visualisées  

prix_ord=[39,25,48,78,66,100]
prix_apple=[48,55,66,70,60,82]
annee=[2015,2016,2017,2018,2019,2020]
```

```
# visualisation des données 
plt.plot(annee, prix_ord, label='prix window', c='g')
plt.plot(annee, prix_apple, label= 'prix apple', c='r')
plt.legend()
plt.xlabel('annees')
plt.ylabel('prix des ordinateur')
plt.title('Evolution des prix des ordinateur durant les 5 dernière années')
```

```
# visualisation des données  avec des nuage de point 
plt.scatter(annee, prix_ord, label='prix window', c='g')
plt.plot(annee, prix_apple, label= 'prix apple', c='r')
plt.legend()
plt.xlabel('annees')
plt.ylabel('prix des ordinateur')
plt.title('Evolution des prix des ordinateur durant les 5 dernière années')
```

```
# visualisation des données  en plusieurs plan
plt.subplot(2,2,1) 
plt.scatter(annee, prix_ord, label='prix window', c='g')
plt.subplot(2,2,4) 
plt.plot(annee, prix_apple, label= 'prix apple', c='r')
```

```
# visualiser les données en diagramme de bande 

index_annee= np.arange(len(annee))
a = 0.60
plt.bar(index_annee+a,prix_ord, label='prix window', width=0.5)
plt.bar(index_annee+a,prix_apple, label='prix apple',width=0.5)
plt.legend()

plt.title('Evolution du prix des ordinateur durant les annees ')
plt.xlabel('Annees')
plt.ylabel('Prix des Ordinateur')
```

```
# Dessiner un pie chart  ou diagramme en secteur 
produit=['p1','p2','p3','p4']
vente=[15,25,50,30]
plt.pie(vente,labels = produit)
```
